```python
df['환불금액'].replace('-', 0, inplace=True)
df['환불금액'] = df['환불금액'].astype('int')
```

```python
# 가장 많이 팔린 코스 상품의 강의 건수를 for문으로 뽑았음 
for course in df['코스(상품) 이름'].unique():
    count = len(df[df['코스(상품) 이름'] == course])
    print(f"'{course}' 강의의 구매 건수: {count}")
```

```python
# 그래프로 그려봄 
category_counts = {}
for category in df['카테고리']:
    if category in category_counts:
        category_counts[category] += 1
    else:
        category_counts[category] = 1

# bar plot 그리기
plt.bar(category_counts.keys(), category_counts.values())
plt.xticks(rotation=45)  # x축 레이블 각도 조정
plt.xlabel('Category')
plt.ylabel('Purchase Count')
plt.show()
```
