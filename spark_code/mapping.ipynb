```python
import pandas as pd
import folium
```

```python
#매핑을위해 csv파일을 불렁모
df_aws = pd.read_csv("dataset/META/awsmap.csv")
df_pm = pd.read_csv("dataset/META/pmmap.csv")
```

```python
#위도와 경도의 평균값을 사용하여 초기 지도의 중심을 설정 
center=df_aws['Latitude'].mean(),df_aws["Longitude"].mean()
#이걸 지도생성
aws_map=folium.Map(location=center,zoom_start=9)

#aws관측소 각 행의 위도와 경도 데이터를 사용하여 지도 마커 추가
for index,row in df_aws.iterrows():
    folium.Marker(
    location=[row['Latitude'], row["Longitude"]],
    popup=f"{row['Location']} {row['Latitude']}, {row['Longitude'],{row['Description']}}",
        icon=folium.Icon(icon='cloud')
    ).add_to(aws_map)
#pm관측소 위도와 경도 데이터를 빨간색 마커로 표기
for index, row in df_pm.iterrows():
    folium.Marker(
    location=[row["Latitude"],row["Longitude"]],
    popup=f"{row['Location']} {row['Latitude']}, ({row['Longitude']},{row['Description']})",
    icon=folium.Icon(color='red')  #folium.Icon으로 색상변경등 아이콘 변경가능
     ).add_to(aws_map)
    
#지도출력
aws_map
```
